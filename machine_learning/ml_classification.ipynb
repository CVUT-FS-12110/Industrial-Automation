# Example of classification - assigning color label to RGB code

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm, tree
```

## Dataset #1 - machine created variations of basic colors

```python
COLORS = {
    "red": [100, 0, 0],
    "blue": [0, 0, 100],
    "green": [0, 100, 0],
    "yellow":  [100, 100, 0],
    "magenta": [100, 0, 100],
    "cyan": [0, 100, 100],
}

training_codes = []
training_labels = []

for key, value in COLORS.items():
    for shade in range(1,11):
        new_value = np.array(value) / 10 * shade
        training_codes.append(new_value)
        training_labels.append(key)
        
for label, code in zip(training_labels, training_codes):
    image = np.ones((20, 200, 3)) * code / 100

    plt.imshow(image)
    plt.title("{} - {}".format(code, label))
    plt.show()
```

```python
clf_tree = tree.DecisionTreeClassifier()
clf_tree.fit(training_codes, training_labels)
plt.figure(figsize=(15,15))
tree.plot_tree(clf_tree)
plt.show()
```

```python
clf_svm = svm.SVC()
clf_svm.fit(training_codes, training_labels)
```

```python
testing_codes = np.random.randint(0, 100, (100, 3))

labels_tree = clf_tree.predict(testing_codes)
labels_svm = clf_svm.predict(testing_codes)

for label_tree, label_svm, code in zip(labels_tree, labels_svm, testing_codes):
    image = np.ones((20, 200, 3)) * code / 100

    plt.imshow(image)
    plt.title("{} - Tree: {}, SVM: {}".format(code, label_tree, label_svm))
    plt.show()
```
