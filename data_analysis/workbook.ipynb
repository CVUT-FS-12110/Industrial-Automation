## Single time series
- find min, max, range, mean, std, median
- remove abnormaliy high/low values from data
- z-score the data

```python
N = 200
t = np.cumsum(np.random.random(N) * 3)
x = 2 * np.sin(t /  (3 * 2 * np.pi)) + 0.5 * np.sin(t /  (0.74 * 2 * np.pi)) + np.linspace(0, 1, N) + np.random.normal(0, 0.2, N)
x[[100, 125, 173]] = 10

plt.plot(t, x)
plt.xlabel("Time [-]")
plt.ylabel("x [-]")
plt.show()
```

## Correlation of two time series

* find a lag between two signals via maximum correlation

```python
N = 10000
x = 5 * np.sin(np.linspace(0,20,N)) + np.random.normal(0,1,N)
y = 5 * np.sin(np.linspace(5,25,N)) + np.random.normal(0.5,0.5,N)

plt.plot(x)
plt.plot(y)
plt.xlabel("Time [-]")
plt.ylabel("x [-]")
plt.show()
```

## Time series regression

- fit coefficents of time series $y = a * (1 + \exp(b * x + c))$
- use the equation to get filtered values for provided $x$ values
- predict value at $x=60$ 

```python
N = 50
x = np.linspace(0, 5, N)
y = 2 + 3 * np.exp(x * 0.8 + 1) + np.random.normal(0, 20, N)

plt.plot(y, "b.")
plt.xlabel("x [-]")
plt.ylabel("y [-]")
plt.show()
```
